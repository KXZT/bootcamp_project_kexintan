```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
sns.set(context='talk', style='whitegrid')
np.random.seed(8)
pd.set_option('display.max_columns', 100)

# === Synthetic data generator (adapt or replace with your own data) ===
n = 2000

df = pd.DataFrame({
    'application_date': pd.date_range('2021-01-01', periods=n, freq='D'),
    'business_age': np.random.randint(1, 20, size=n),  # Years in business
    'annual_revenue': np.random.randint(50000, 2000000, size=n),  # Annual revenue in dollars
    'credit_score': np.random.randint(500, 850, size=n),  # Credit score
    'loan_amount': np.random.randint(10000, 500000, size=n),  # Loan amount requested
    'existing_debt': np.random.randint(0, 300000, size=n),  # Existing business debt
})

df['debt_to_income_ratio'] = (df['existing_debt'] / df['annual_revenue']).round(2)
df['loan_to_income_ratio'] = (df['loan_amount'] / df['annual_revenue']).round(2)

default_prob = 0.4 * (850 - df['credit_score']) / 350 + 0.3 * df['debt_to_income_ratio'] + 0.2 * df['loan_to_income_ratio'] + -0.1 * (df['business_age'] / 20) + np.random.normal(0, 0.1, size=n) 
default_prob = np.clip(default_prob, 0, 1)
df['default_probability'] = default_prob.round(3)
df['default'] = (df['default_probability'] > np.random.uniform(0, 0.5, size=n)).astype(int)
df['industry'] = np.random.choice(['Retail', 'Services', 'Manufacturing', 'Other'], size=n)


# inject a bit of missingness and outliers
df.loc[np.random.choice(df.index, 5, replace=False), 'credit_score'] = np.nan
df.loc[np.random.choice(df.index, 3, replace=False), 'business_age'] = np.nan
df.loc[np.random.choice(df.index, 2, replace=False), 'loan_amount'] = df['loan_amount'].max()+200000
df.to_csv('small_business_loans.csv', index=False)

df.head()
```

## 1) First look

```python
df.info(), df.isna().sum()
```

## 2) Numeric profile

```python
desc = df[['credit_score', 'loan_amount', 'annual_revenue', 'existing_debt', 'business_age', 'debt_to_income_ratio', 'loan_to_income_ratio']].describe().T
desc['skew'] = [skew(df[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(df[c].dropna()) for c in desc.index]
desc
```

## 3) Distributions (TODO: add at least 3)

```python
# TODO: histogram + KDE for a key variable
plt.figure(figsize=(10, 6))
sns.histplot(df['credit_score'].dropna(), kde=True, bins=30)
plt.title('Credit Score Distribution')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['loan_amount'])
plt.title('Loan Amount Distribution (Outliers)')
plt.xlabel('Loan Amount ($)')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['debt_to_income_ratio'].dropna(), kde=True, bins=30)
plt.title('Debt-to-Income Ratio Distribution')
plt.xlabel('Debt-to-Income Ratio')
plt.ylabel('Frequency')
plt.show()
```

## 4) Relationships (TODO: add at least 2)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='credit_score', y='default_probability', alpha=0.6)
plt.title('Credit Score vs Default Probability')
plt.xlabel('Credit Score')
plt.ylabel('Default Probability')
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='debt_to_income_ratio', y='default_probability', hue='industry', alpha=0.6)
plt.title('Debt-to-Income Ratio vs Default Probability (by Industry)')
plt.xlabel('Debt-to-Income Ratio')
plt.ylabel('Default Probability')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## 5) (Optional) Correlation matrix

```python
corr = df[['age','income','transactions','spend']].corr(numeric_only=True)
sns.heatmap(corr, annot=True, fmt='.2f', cmap='vlag', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
corr
```
